```python
#import ....
import pandas as pd
import numpy as np
food =  pd.read_csv("food_nutrient_2011_13_AHS.csv", header=0,low_memory=False)
```

```python
#Question 1
#print the number of food, number of attributes, attribute names and their datatype. 

print("***")
print("Q1")
#print the number of food
numOfFood = food.shape[0]
print("Number of foods:", numOfFood)

#print the number of attribute 
numOfAttr = food.shape[1]
print("Number of attributes:", numOfAttr)
print("")

#print the type of the attribute 
print(food.dtypes)
print("***")
```

```python
#Question 2 Basic Statistics 
# What is the median value of Magnesium(Mg)(mg)? What is the mean value of Moisture(g)? 

print("***")
print("Q2")
#print he median value of Magnesium(Mg)(mg)
medianOfMag = food["Magnesium (Mg) (mg)"].median()
print( "Median value of Magnesium:", round(medianOfMag, 1))

#print the mean value of Magnesium(g)
meanOfMoi = food["Moisture (g)"].mean()
print("Mean value of Moisture:",round(meanOfMoi,1))
print("***")
```

```python
#Question 3

# Part a: Convert the datatype of the Survey ID attribute into the string type
food['Survey ID'] = food['Survey ID'].astype('str')


#Part b: Using Survey ID, create a new attribute Food catefory, which contains the first two digits of the attrubute Survey ID. 

#create the new attribute according to the survey ID
foodCate = food['Survey ID'].str.slice(0, 2)
#insert the attribute back to the dataframe
food["Food category"] = foodCate


# Part c: Using this new attribute, compute the number of foods with Food category 
# equal to '13'(this category can be described as "Ceneal vased products and dishes")
# Using this information, now print out the percentage of foods which are in this category

#find the number of food that food category is 13
numOf13 = food[food["Food category"]=="13"].shape[0]
#calculate the percentage
percentage = numOf13/numOfFood*100
#print out the percentage
print("***")
print("Q3")
print("% of foods which are Cereal based products and dishes (Food category 13) =", round(percentage, 1))
print("***")
```

```python
#Question 4
#draw a plot consisting of two boxplot. 
#One boxplot to show tge distribution of Total fat(g) within foods that are"Cereal based products and dishes"
#the othter boxplot to show the distribution of Protein(g)within foods that are"Cereal based products and dishes"


#import the matplotlib 
#and this library will also being used by the following question

import matplotlib.pyplot as plt
%matplotlib inline

#find the data for the Total fat
boxdata1 = food[food["Food category"]=="13"]["Total fat (g)"]
#find the data for the Protein 
boxdata2 = food[food["Food category"]=="13"]["Protein (g)"]
#create the boxplot
newDataFrame = pd.DataFrame({'Total fat(g)':boxdata1, 'Protein (g)':boxdata2})
#print out the boxplot
newDataFrame.plot.box()

plt.title("Boxplot of Total Fat and Protein")
```

```python
#Question 5
#Consider the Food Category attribute creates in Question 3. Draw a bar plot showing Food category(x-axis) versus mean Total suger(g)(y-axis)

#group by the data by the food category and find the mean of total sugars(g)
barinformation = food.groupby('Food category')['Total sugars (g)'].mean()
barinformation.plot(kind='bar')

#and add a label for the barplot
plt.ylabel("mean Total sugars (g)")
plt.title("Average Sugar Content of Various Food Category")
```

```python
#Question 6 Scatter plot

#Part a: Create an attribute EnergyLevel which has value “1”
food['EnergyLevel'] = food['Energy, with dietary fibre (kJ)'].apply(lambda x: '1' if x > 1000 else '0')

#Part b:Draw a scatter plot showing Total sugers(g)(x-axis) versus Protein(g)(y-axis) for all foods A food should have red color if it has "High"

#seperate that value according to the energylevel
blueElement = food[food["EnergyLevel"]=='0']
redElement = food[food["EnergyLevel"]=='1']

#plot one scatter plot first and add another data at the same plot
ax = blueElement.plot.scatter(x="Total sugars (g)",y="Protein (g)", c='b', label='Low energy')
redElement.plot.scatter(x="Total sugars (g)",y="Protein (g)",c='r', label = 'High energy', ax=ax)
plt.title("Sugar and Protein Content of Foods with Different Level Energies")

```

```python
#Question 7

#Part a

#find the maximum and minimum value of Protein
proteinMax = food["Protein (g)"].max()
proteinMin = food["Protein (g)"].min()

#find the maximum and minimum value of Total fat
fatMax = food["Total fat (g)"].max()
fatMin = food["Total fat (g)"].min()

#find the maximum and minumum value of Total sugur
sugerMax = food["Total sugars (g)"].max()
sugerMin = food["Total sugars (g)"].min()

#create new attribute by normalizing the Protein, Total fat ane Total sugar in the dataframe 
food["Protein (g)"] = food["Protein (g)"].map(lambda x: (x-proteinMin)/(proteinMax-proteinMin))
food["Total fat (g)"] = food["Total fat (g)"].map(lambda x: (x-fatMin)/(fatMax-fatMin))
food["Total sugars (g)"] = food["Total sugars (g)"].map(lambda x: (x-sugerMin)/(sugerMax-sugerMin))


#Part b

#impor the lib for parallel_coordinates
from pandas.plotting import parallel_coordinates

#create the attribute energylevel depends on the attribute 'Energy, with dietary fibre (kJ)'
food['EnergyLevel'] = food['Energy, with dietary fibre (kJ)'].apply(lambda x: 'High energy' if x > 1000 else 'Low energy')

#create an dataframe consist with the element need to use for the parallel_coordinate plot
col_n = ['EnergyLevel',"Total sugars (g)","Total fat (g)", "Protein (g)"]
graph = pd.DataFrame(food,columns = col_n)

#plot the paralle coordinate graph
colors = {'High energy':'r', 'Low energy':'b'}
parallel_coordinates(graph[["Total sugars (g)","Total fat (g)", "Protein (g)","EnergyLevel"]],'EnergyLevel', color = graph['EnergyLevel'].apply(lambda x:colors[x]))
plt.title("Sugar, Fat and Protein Content of Foods with Different Level Energies")
plt.show()
```

```python
#Question 8

#Part a
#impor the csv file again
food =  pd.read_csv("food_nutrient_2011_13_AHS.csv", header=0,low_memory=False)

#create the attribute'calorie_count_per_100g' by  normalizing the value of attribute protein, available carbohydrates with sugar alcohols, Total fat and Alcohol  
food['calorie_count_per_100g'] = food.apply(lambda x:  4 * x["Protein (g)"] + 4 * x['Available carbohydrates, with sugar alcohols (g)'] + 9*x["Total fat (g)"]+7*x['Alcohol (g)'] , axis=1)


#Part b
#sort the value of food by the value of attribute 'calorie_count_per_100g' in descending order 
food = food.sort_values( by = 'calorie_count_per_100g',ascending = False)

#take first 5 element of food dataframe (which is top 5 food with the highest'calorie_count_per_100g')
top5Calorie = food.iloc[0:5]

#print the first 5 element out
print("***")
print("Q8")
print("")
for i in range(5):
    print (str(i+1)+'.',top5Calorie.iloc[i]["Food Name"],top5Calorie.iloc[i]['calorie_count_per_100g'])
print("***")
    

    
    
# part c


#import the matplotlib colors
import matplotlib.colors as colors


#create the attribute "Food category" again, code is copy from question 3
food['Survey ID'] = food['Survey ID'].astype('str')
foodCate = food['Survey ID'].str.slice(0, 2)
food["Food category"] = foodCate

#groupby the value and polt it's mean in the pie plot 
piegraph = food.groupby('Food category')['calorie_count_per_100g'].mean()

#store all the possible color in the list 
colors_list = list(colors._colors_full_map.values())

#create the pie plot 
plt.pie(piegraph.values, explode = None, labels = piegraph.index,colors = colors_list, autopct = '%.1f%%')

#add a title
plt.title("Average Calorie Distribution of Different Food category")

#make the the pie plot a circle pie
plt.axis('equal')

#print the pie on the screen
plt.show()

```

```python
#Question 9

#prepare all the dataframe at first

#import the "food_nutrient_2011_13_AHS.csv" again
food =  pd.read_csv("food_nutrient_2011_13_AHS.csv", header=0,low_memory=False)
#import "8i. Australian Health Survey Classification System.csv" name with foodcat
foodcat =  pd.read_csv("8i. Australian Health Survey Classification System.csv", header=0,low_memory=False)

#create the attribute "Food category" agian, code is copy from question3
food['Survey ID'] = food['Survey ID'].astype('str')
foodCate = food['Survey ID'].str.slice(0, 2)
food["Food category"] = foodCate



#Part a

#convert the data type of 'Food Group Code' into the string type
foodcat["Food Group Code"] = foodcat["Food Group Code"].astype(str)

#flitter the "Food Group Code", only the string with the length 2
foodcat= foodcat[foodcat["Food Group Code"].str.len()==2]

#change the name of the attribute so that can merge two the dataframe together
foodcat.columns = ["Food category", "Food category name"]

#merge dataframe food and foodcat together 
food = pd.merge(food,foodcat,on="Food category",how='inner')



#Part b
#import the json library
import json
#create a series food_suger store food category name and the mean of total suger after groupby of food category name
food_suger = food.groupby("Food category name")["Total sugars (g)"].mean()
food_suger.columns = ["Food category", "Mean Total Suger"]

#the data has already prepared, we need to change it into the json form

#create a dictionary to store all the series data
foodDic = {}

#fill the dictionary by the for loop
for i in food_suger.index:
    foodDic[i] = round(food_suger[i], 1)

#add name for the dictionary
jsonDic={"Mean total suger(g) by category":foodDic}


#change the dictionary into the json form string
j_str = json.dumps(jsonDic)

#load the string to the json format
d1 = json.loads(j_str)
d1
```
