```python
import pandas as pd
import numpy as np
wine =  pd.read_csv("datasets/wine.data_copy", header = 0)
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
wine
```

```python
plt.figure()
for index, row in wine.iteritems(): 
    plt.hist(wine[index], bins = 20,color = 'steelblue', edgecolor = 'k', label = 'historgram') 
    plt.title(index)
    plt.show()
    
```
